```python
import cv2
import threading
import tkinter as tk
from PIL import Image, ImageTk


def start_capture():
    global capture, last_frame

    capture = cv2.VideoCapture(0)
    
    fourcc = cv2.VideoWriter_fourcc("X", "V", "I", "D")
    video_writer = cv2.VideoWriter(r"sample.avi", fourcc, 30.0, (640, 480))

    while running:
        rect, frame =  capture.read()

        if rect:
            cv2image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA)
            last_frame = Image.fromarray(cv2image)
            video_writer.write(frame)

    capture.release()
    video_writer.release()
    
def start_capture_():
    global capture, last_frame
    capture = cv2.VideoCapture(0)
    
    while running:
        rect, frame =  capture.read()
        if rect:
            cv2image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA)
            last_frame = Image.fromarray(cv2image)
            
    capture.release()


def update_frame():
    if last_frame is not None:
        tk_img = ImageTk.PhotoImage(master=video_label, image=last_frame)
        video_label.config(image=tk_img)
        video_label.tk_img = tk_img

    if running:
        root.after(10, update_frame)

def start_rec():
    global running

    running = True
    thread = threading.Thread(target=start_capture, daemon=True)
    thread.start()
    update_frame()

def start_rec_():
    global running

    running = True
    thread = threading.Thread(target=start_capture_, daemon=True)
    thread.start()
    update_frame()

def start_rec__(event):
    global running

    running = True
    thread = threading.Thread(target=start_capture, daemon=True)
    thread.start()
    update_frame()

def snapshot():
    start_rec_()
    global capture
    capture = cv2.VideoCapture(0)
    
    while running:
        rect, frame =  capture.read()
        cv2.imshow('my_image', frame)
        stop_rec()
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    capture.release()
    
def func4():
    global count
    count = count + 1
   
    if count%2 == 1:
        print("Start Recording...")
        print("Warming up camera...")
        start_rec_()
    else:
        print("Stop Recording...")
        stop_rec()
        
def func3():
    global count
    count = count + 1
   
    if count%2 == 1:
        print("Start Recording...")
        print("Warming up camera...")
        print("Saving video...")
        start_rec()
    else:
        print("Stop Recording...")
        stop_rec()   

# Dung stream camera 
def stop_rec():
    global running
    running = False
    
def stop_rec_(event):
    global running
    running = False
    
def closeWindow():
    stop_rec()
    root.destroy()


running = False
after_id = None
last_frame = None
count = 0

root = tk.Tk()
root.protocol("WM_DELETE_WINDOW", closeWindow)

video_label = tk.Label()
video_label.pack(expand=True, fill="both")

start_button1 = tk.Button(text="Button 1", command=snapshot)
start_button1.pack()

start_button2 = tk.Button(text="Button 2")
start_button2.pack()
start_button2.bind('<ButtonPress-1>',start_rec__)
start_button2.bind('<ButtonRelease-1>',stop_rec_)

start_button3 = tk.Button(text="Button 3", command=func3)
start_button3.pack()

start_button4 = tk.Button(text="Button 4", command=func4)
start_button4.pack()

root.mainloop()
```
