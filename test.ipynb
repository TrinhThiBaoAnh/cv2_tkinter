```python
import cv2
import threading
import tkinter as tk
from PIL import Image, ImageTk
import numpy as np

def preprocess(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        
    # Create first mask
    lower_red = np.array([0,150,50])
    upper_red = np.array([5,255,255])
    # Threshold the HSV image to get only red colors
    mask = cv2.inRange(hsv, lower_red, upper_red)
     
    # Create second mask
    lower_red2 = np.array([175,150,50])
    upper_red2 = np.array([179,255,255])
    # Threshold the HSV image to get only red colors
    mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
    
    #combine masks
    final_mask = cv2.bitwise_or(mask, mask2)
        
    # apply mask to original image
    result = cv2.bitwise_or(img,img, mask= final_mask)
    
    return final_mask, result

def start_capture_():
    global capture, last_frame
    capture = cv2.VideoCapture(0)
    global running
    while running:
        ret, img = capture.read()
        img = cv2.imread('a.png')
        final_mask, result = preprocess(img)
        
        contours, hierarchy = cv2.findContours(final_mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        for i in range(len(contours)):
            cv2.drawContours(result, contours, i, (255,255,255), 2, cv2.LINE_8, hierarchy, 0)
            
        last_frame = Image.fromarray(result)
        
        cv2.imshow('image', img)
        key = cv2.waitKey(30)
        if key == ord('q') or key == 27:
            cv2.destroyAllWindows()
            running = False
            break
            

def start_capture():
    global capture, last_frame
    capture = cv2.VideoCapture(0)
    global running
    while running:
        ret, img = capture.read()
        img = cv2.imread('a.png')
        final_mask, result = preprocess(img)
        last_frame = Image.fromarray(result)
        #show image
        cv2.imshow('image', img)
        key = cv2.waitKey(30)
        if key == ord('q') or key == 27:
            cv2.destroyAllWindows()
            running = False
            break

def update_frame():
    if last_frame is not None:
        tk_img = ImageTk.PhotoImage(master=video_label, image=last_frame)
        video_label.config(image=tk_img)
        video_label.tk_img = tk_img

    if running:
        root.after(10, update_frame)

def start_rec():
    global running
    running = True
    thread = threading.Thread(target=start_capture, daemon=True)
    thread.start()
    update_frame()

def start_rec_():
    global running
    running = True
    thread = threading.Thread(target=start_capture_, daemon=True)
    thread.start()
    update_frame()   

def stop_rec():
    global running
    running = False
    
def closeWindow():
    stop_rec()
    root.destroy()
    
running = False
last_frame = None

root = tk.Tk()
root.protocol("WM_DELETE_WINDOW", closeWindow)

video_label = tk.Label()
video_label.pack(expand=True, fill="both")

start_button1 = tk.Button(text="Button 1", command=start_rec)
start_button1.pack()

start_button3 = tk.Button(text="Button 2", command=start_rec_)
start_button3.pack()


root.mainloop()
```
